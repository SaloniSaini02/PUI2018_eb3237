## Getting data

```python
API_URL = 'https://fb55.carto.com:443/api/v2/sql?q='

def queryCartoDB(query, formatting = 'CSV', source = API_URL):
    '''queries carto datasets from a given carto account
    Arguments: 
    query - string: a valid sql query string
    format - outlut format  OPTIONAL (default CSV)
    source - a valid sql api endpoint OPTIONAL (default carto fb55 account)
    Returns:
    the return of the sql query AS A STRING
    NOTES:
    designed for the carto API, tested only with CSV return format'''
    
    
    data = urlencode({'format': formatting, 'q': query}).encode("utf-8")
    try:
        response = urlopen(source, data)
        return response.read()
    except HTTPError as e:
        raise (ValueError('\n'.join(ast.literal_eval(e.readline())['error'])))
        
def get_data(query):
    try:
        return pd.read_csv(io(queryCartoDB(query)), sep = ',')
    except ValueError as v:
        print (str(v))
```

### Task 2 — Working with date/time

```python
#Selecting trips started on Feb-02-2015 only 
query = '''SELECT * FROM fb55.citibike
WHERE starttime::date = '2015-02-02'
'''

table = get_data(query)
table.head(5)
```

## Task 3 — Working with Space

```python
#Showing the list of start station locations & names
#Using GROUP BY

#limit by 10

query = '''SELECT start_station_name, start_station_longitude, start_station_latitude FROM fb55.citibike
GROUP BY start_station_name, start_station_longitude, start_station_latitude
LIMIT 10'''

table = get_data(query)
table
```

```python
# Showing the number of trips started per station - all stattions
#using start_station_id - not name

query = '''SELECT start_station_id, COUNT(bikeid) as bike FROM fb55.citibike
GROUP BY start_station_id
ORDER BY bike DESC
'''
table = get_data(query)
table.head()
```

```python
#Showing the number of trips started per station
#but only for stations within 500m of Time Square!

#using start_station_name and coordinates

#The coordinates of Time Square is (40.7577,-73.9857)

#formula:
#Tangent line to circle (x − x0)^2 + (y − y0)^2 = r^2
#through point P (a, b)
#Point O (x0, y0) is the center of the circle.

#in Our caase:
#(x - x0)^2 + (y - y0)^2 <= 0,0045045045045045045045045045045^2

#0,0045045045045045045045045045045^2 is radius squared (500 meters) translated in latitude/longtitude

query = '''
    SELECT start_station_name
        , start_station_longitude
        , start_station_latitude
        , COUNT(bikeid) as bike 
    FROM fb55.citibike
    WHERE POWER(start_station_latitude - 40.7577,2) + POWER(start_station_longitude + 73.9857, 2) <= 0.000020290560831101371641912182452723
    GROUP BY start_station_name, start_station_longitude, start_station_latitude
    ORDER BY bike DESC
'''

table = get_data(query)
table
```

## Task 4 — Putting it all together

```python
#Find the station that had the longest average trip duration 
#during weekends and within 500m of TimeSquare! 

query = '''
    SELECT start_station_name
            , start_station_longitude
            , start_station_latitude
            , AVG(tripduration) as average
    FROM fb55.citibike
    WHERE POWER(start_station_latitude - 40.7577,2) + POWER(start_station_longitude + 73.9857, 2) <= 0.000020290560831101371641912182452723
    AND EXTRACT (DAY FROM starttime::date) IN (0,1)
    GROUP BY start_station_name, start_station_longitude, start_station_latitude
    ORDER BY average DESC
'''

table = get_data(query)
table

#The first station in the table has the longest average trip duration
```

```python
#alternatively, could be query within a query to show only that station

query = '''
    with t as (
      SELECT start_station_name
            ,start_station_longitude
            ,start_station_latitude
            ,AVG(tripduration) as average
        FROM fb55.citibike
       WHERE POWER(start_station_latitude - 40.7577,2) + POWER(start_station_longitude + 73.9857, 2) <= 0.000020290560831101371641912182452723
         AND EXTRACT (DAY FROM starttime::date) IN (0,1)
       GROUP BY start_station_name, start_station_longitude, start_station_latitude
    )
    select * from t where average = (select max(average) from t)
'''
table = get_data(query)
table
```
